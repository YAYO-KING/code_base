```python
# 列表List

# 创建列表
list1 = ['yanyue', 29, '12443vd@qq.com', '2020-02-09']
list2 = [5, 6, 7, 8, 1, 2, 3, 4]
print(list1)  # ['yanyue', 29, '12443vd@qq.com', '2020-02-09']

# 列表取值
# 下标取值
print(list1[0])  # yanyue
print(list1[-1])  # 12443vd@qq.com
# 范围取值(包前不包后)
listTemp = list1[1:3]
print(listTemp)  # [29, '12443vd@qq.com']

# index(xxx) 用于获取指定元素的索引值
print(list1.index(29))  # 1

# for循环 forin  遍历列表
for item in list1:
    if item == 'yanyue':
        print(item)

# len() 获取列表的长度
print(len(list1))  # 4

# list.reverse() 列表的翻转 注意：本身没有返回值的
list1.reverse()
print(list1)

# list.sort() 列表中元素进行排序
list2.sort(reverse=True)
print(list2)

# 列表新增元素
# append
list1.append('YAYO')  # 列表末尾追加新元素
print(list1)  # ['2020-02-09', '12443vd@qq.com', 29, 'yanyue', 'YAYO']
# insert(index,obj)
list1.insert(2, 'KOKO')  # 在指定索引位置插入新的元素
print(list1)  # ['2020-02-09', '12443vd@qq.com', 'KOKO', 29, 'yanyue', 'YAYO']

# 列表修改元素
list1[3] = 40  # 修改指定索引位置的数据
print(list1)  # ['2020-02-09', '12443vd@qq.com', 'KOKO', 40, 'yanyue', 'YAYO']
list1[1:3] = ['yanyue']  # 更新指定范围的数据
print(list1)  # ['2020-02-09', 'yanyue', 40, 'yanyue', 'YAYO']

# 列表删除元素
list1.remove('yanyue')  # 删除指定元素
print(list1)  # ['2020-02-09', 40, 'yanyue', 'YAYO']
list1.pop(2)  # 按索引删除指定元素
print(list1)  # ['2020-02-09', 40, 'YAYO']

# count 获取列表中某个元素出现的次数
count = list1.count('YAYO')
print(count)  # 1

# append 将整个列表追加到末尾
list1.append(['哈哈哈', 'Oh my God'])
print(list1)  # ['2020-02-09', 40, 'YAYO', ['哈哈哈', 'Oh my God']]
# extend 将列表中的元素追加到原始列表末尾
list1.extend(['哈哈哈', 'Oh my God'])
print(list1)  # ['2020-02-09', 40, 'YAYO', ['哈哈哈', 'Oh my God'], '哈哈哈', 'Oh my God']

# in 判断元素是否在列表中
print('哈哈哈' in list1)  # True

# remove删除列表中的一个元素
list1.remove('哈哈哈')

# copy 复制一个列表
listCopy = list1.copy()
list3 = list1
print(listCopy)  # ['2020-02-09', 40, 'YAYO', ['哈哈哈', 'Oh my God'], 'Oh my God']
print(list1 is list3)  # True
print(listCopy is list3)  # False

# clear清空列表
list1.clear()
print(list1)  # []
print(list3)  # []
print(listCopy)  # ['
```
