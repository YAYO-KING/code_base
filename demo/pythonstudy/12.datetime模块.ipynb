```python
 1.datetime模块的使用
# import datetime
from datetime import datetime, date, time, timedelta

# 2.time 时间模块
# import time

print(dir(datetime))
# ['MAXYEAR', 'MINYEAR', '__builtins__', '__cached__', '__doc__', '__file__', '__loader__', '__name__', '__package__', '__spec__', 'date', 'datetime', 'datetime_CAPI', 'sys', 'time', 'timedelta', 'timezone', 'tzinfo']

# datetime.datetime.now() 获取当前时间
nowTime = datetime.now()  # 2020-09-19 21:24:22.675923
print("now is :{0}".format(nowTime))

# 获取当前日期
print(nowTime.date())  # 2020-09-19

# 获取当前时间
print(nowTime.time())  # 21:26:22.793275

# 获取当前时间
print(datetime.today())  # 2020-09-19 21:28:40.858069

# 获取当前时间的year
print(nowTime.year)  # 2020

# 获取当前时间的month
print(nowTime.month)  # 9

# 获取当前时间的day
print(nowTime.day)  # 19

# 获取1970.xxx.xxx到现在的毫秒数
# print(time.time())  # 1600522367.2238314

# 3.DateTime模块转换参数
# 自定义时间和日期
# datetime 可以设置日期和时间
d1 = datetime(2020, 11, 11, 11, 11, 11)
print(d1)  # 2020-11-11 11:11:11
# date 只能设置日期
d2 = date(2020, 12, 12)
print(d2)  # 2020-12-12
# time() 设置时间 分钟 秒
t = time(11, 11, 11)
print(t)  # 11:11:11

# 4.日期、时间和字符串之间的相互转化
dateStr = "2018-08-25 11:11:11"
# 将字符串转换成datetime对象 strptime
date2 = datetime.strptime(dateStr, "%Y-%m-%d %H:%M:%S")  # 2018-08-25 11:11:11
print(date2)
# 将datetime对象转换成字符串 strftime
dateStr2 = nowTime.strftime("%Y-%m-%d %H:%M:%S")
print(dateStr2)  # 2020-09-19 22:21:41

# 5.datetime之间的加减操作
dateNext = nowTime + timedelta(days=5, hours=42, minutes=4, seconds=56, microseconds=444)  # 5天42小时后
print(nowTime)  # 2020-09-19 22:24:23.022639
print(dateNext)  # 2020-09-26 16:24:00.311080

# 时间的减法
d1 = datetime(2018, 10, 15)
d2 = datetime(2018, 11, 12)
rest = d2 - d1
print(rest.days)  # 28
print(dir(rest))
# ['__abs__', '__add__', '__bool__', '__class__', '__delattr__', '__dir__', '__divmod__', '__doc__', '__eq__', '__floordiv__', '__format__', '__ge__', '__getattribute__', '__gt__', '__hash__', '__init__', '__init_subclass__', '__le__', '__lt__', '__mod__', '__mul__', '__ne__', '__neg__', '__new__', '__pos__', '__radd__', '__rdivmod__', '__reduce__', '__reduce_ex__', '__repr__', '__rfloordiv__', '__rmod__', '__rmul__', '__rsub__', '__rtruediv__', '__setattr__', '__sizeof__', '__str__', '__sub__', '__subclasshook__', '__truediv__', 'days', 'max', 'microseconds', 'min', 'resolution', 'seconds', 'total_seconds']
```
