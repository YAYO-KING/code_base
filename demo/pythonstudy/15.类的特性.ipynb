```python
class Cat:
    """猫科动物"""

    # 类属性
    tag = "我是家猫"

    def __init__(self, name, age, sex=None):
        # 实例属性
        self.name = name
        # __xxx 私有属性，外部无法直接访问
        self.__age = age
        self.sex = sex

    def set_age(self, age):
        """
        修改猫的年龄
        :param age:
        :return:
        """
        self.__age = age

    def show_info(self):
        """
        显示猫的信息
        :return:
        """
        result = "我叫{0}，今年{1}岁了,性别：{2}".format(self.name, self.__age, self.sex)
        print(result)
        return result

    def eat(self):
        """吃"""
        print("猫喜欢吃鱼")

    def catch(self):
        """捉老鼠"""
        print("猫喜欢捉老鼠")


# 当.py文件被直接运行时，if __name__ == '__main__'之下的代码块将被运行；
# 当.py文件以模块形式被导入时，if __name__ == '__main__'之下的代码块不被运行
if __name__ == "__main__":
    # 实例化一个对象
    tomcat = Cat("Tom", 10, "公的")
    tomcat.eat()
    tomcat.show_info()  # 我叫Tom，今年10岁了,性别：公的
    print(tomcat.name)
    # print(tomcat.age) 无法获取到 因为age是类是私有属性。无法直接通过类去访问
    # print(tomcat.__age)
    tomcat.set_age(15)
    tomcat.show_info()  # 我叫Tom，今年15岁了,性别：公的
    # 类属性  对象和类本身都可以直接访问
    print(tomcat.tag)  # 我是家猫
    print(Cat.tag)  # 我是家猫

    # isinstance 判断对象是否是类的实例
    print(isinstance(tomcat, Cat))  # True
```
