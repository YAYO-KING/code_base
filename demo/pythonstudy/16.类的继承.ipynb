```python
class BaseCat:
    """猫科动物 基类"""
    tag = "猫科动物"

    def __init__(self, name):
        self.name = name

    def eat(self):
        print("猫都要吃东西")


class Tiger(BaseCat):
    def eat(self):
        # 调用父类中的方法
        super(Tiger, self).eat()
        # 扩充下自己的方法内容
        print("我还喜欢吃大肉")

class Panda(BaseCat):
    def eat(self):
        # 调用父类中的方法
        super(Panda, self).eat()
        # 扩充下自己的方法内容
        print("我还喜欢吃竹子")


class PetCat(BaseCat):
    def eat(self):
        # 调用父类中的方法
        super(PetCat, self).eat()
        # 扩充下自己的方法内容
        print("我还喜欢吃猫粮")


if __name__ == "__main__":
    tiger = Tiger("老虎")
    tiger.eat()
    panda = Panda("熊猫")
    panda.eat()
    petcat = PetCat("宠物猫")
    petcat.eat()

    print(issubclass(Tiger, BaseCat))  # True
    print(issubclass(Tiger, Panda))  # False
```

```python
# 类的多重继承

class BaseCat:
    """猫科动物 基类"""
    tag = "猫科动物"

    def __init__(self, name):
        self.name = name

    def eat(self):
        print("猫都要吃东西")


class ProtectedMixin:
    """国家保护动物类"""

    def protected(self):
        print("国家保护动物")


class Tiger(BaseCat, ProtectedMixin):  # 类可以多继承的
    # 注意：继承多个类的时候，如果类中的方法存在同名的情况，会只触发先继承的类的方法
    def eat(self):
        # 调用父类中的方法
        super(Tiger, self).eat()
        # 扩充下自己的方法内容
        print("我还喜欢吃大肉")


class Panda(BaseCat, ProtectedMixin):
    def eat(self):
        # 调用父类中的方法
        super(Panda, self).eat()
        # 扩充下自己的方法内容
        print("我还喜欢吃竹子")


class PetCat(BaseCat):
    def eat(self):
        # 调用父类中的方法
        super(PetCat, self).eat()
        # 扩充下自己的方法内容
        print("我还喜欢吃猫粮")


if __name__ == "__main__":
    tiger = Tiger("老虎")
    tiger.eat()
    tiger.protected()
    panda = Panda("熊猫")
    panda.eat()
    panda.protected()
    petcat = PetCat("宠物猫")
    petcat.eat()

    print(issubclass(Tiger, BaseCat))  # True
    print(issubclass(Tiger, Panda))  # False
```
