```python
# 身份证 号码的校验
# 需要使用分组的概念
# 1.前6位：区域的编号
# 2.接下来的4位：出生的日期
# 3.接下来的4位：出生的月份、日期
# 4.最后4位：
# 5.倒数第二位：性别

# 身份证
# (^\d{15}$)|(^\d{18}$)|(^\d{17}(\d|X|x)$)

# 邮箱
# ^[\w-]+(\.[\w-]+)*@[\w-]+(\.[\w-]+)+$

# compile方法 -- 生成正则对象
# match方法 -- 比较字符串和正则表达式，match功能有限只能匹配字符串开始的时候是否和正则表达式匹配，匹配不上就是None
# findall方法 -- 查找所有匹配到的文字返回一个列表
# search -- 查看字符串中是否有匹配正则表达式的内容


import re

# 1.将正则表达式编译(compile)成正则表达式的对象  # re.IGNORECASE忽视大小写
pattern = re.compile(r'hello', re.IGNORECASE)

# 2.通过match进行匹配 -- 查询字符串开头是不是有匹配项 world写开头都不行
rest = pattern.match('hello world')
print(rest)  # <re.Match object; span=(0, 5), match='hello'>
print(rest.string)  # hello world
print(rest.re)  # re.compile('hello', re.IGNORECASE)
print(rest.group())  # hello

# 3.findall 找出所有匹配到的字符串并且返回一个列表
data = "chjsdkahcjkds2324njknrjk435345434jkljklj2k3423jkj4k32jk43"
p1 = re.compile(r'\d+')  # 数字类型
# p1 = re.compile(r'[a-z]+', re.I)  # 字母类型
rest1 = p1.findall(data)
print(rest1)  # ['2324', '435345434', '2', '3423', '4', '32', '43']

# 不使用编译的对象
rest2 = re.findall(r'[a-z]+', data, re.I)
print(rest2)  # ['chjsdkahcjkds', 'njknrjk', 'jkljklj', 'k', 'jkj', 'k', 'jk']

# 4.search 方法 -- 查询整个字符串里面是否有匹配项
content = "hello world"
p3 = re.compile(r'world')
rest3 = p3.search(content)
print(rest3)  # <re.Match object; span=(6, 11), match='world'>

rest4 = re.search(r'world', 'hello world', re.I)
print(rest4)  # <re.Match object; span=(6, 11), match='world'>
```

```python
# 正则表达式的分组（group()与groups()）

import re


def test_group():
    content = "hello,world"
    pattern1 = re.compile(r'world')
    rest1 = pattern1.search(content)
    print(rest1)  # <re.Match object; span=(6, 11), match='world'>
    if rest1:
        # 必须要有能匹配上的时候才可以使用group()函数
        print(rest1.group())  # world
        print(rest1.groups())  # ()


def testIdCard(cardNum):
    pattern = re.compile(r'(^\d{15}$)|(^\d{18}$)|(^\d{17}(\d|X|x)$)')
    rest = pattern.search(cardNum)
    if rest:
        # 如果正则里面有分组表达式的话 (xxx)|(xxx)
        # groups() 返回的就是所有匹配到的分组
        print(rest.groups())
        # group(1) 匹配到的第一个分组内容
        print(rest.group(1))  # None
        print(rest.groupdict())  # {}
        return rest.string
    else:
        return None


if __name__ == "__main__":
    # test_group()
    print(testIdCard('320882199305280239'))
```
