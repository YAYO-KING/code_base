```python
# pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
from urllib.parse import urlparse, parse_qs
import requests
import json

# 设置 Chromedriver 的路径
chromedriver_path = 'D:\\YAYO\\chrome\\chromedriver119.exe'

# 配置 Chromedriver 的位置
options = webdriver.ChromeOptions()

# 设置浏览器启动最大化
options.add_argument('--start-maximized')
# 设置 Chrome 为无头模式（即不打开浏览器界面）
options.add_argument("--headless")
# 设置浏览器的 User-Agent
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.140 Safari/537.36 Edge/17.17134")
# 添加代理
# options.add_argument('--proxy-server=http://your-proxy.com:1234')

# 启动浏览器 Google
driver = webdriver.Chrome(executable_path=chromedriver_path, chrome_options=options)

# 打开网页
driver.get("https://search.tiangong.cn/")

# 查看元素
# element = driver.find_element_by_id("element_id")

# 等待元素加载好
element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "input-wrapper"))
)

inputEle = driver.find_element(By.XPATH, '//div[@id="input-wrapper"]//input[1]')

inputEle.send_keys("计算机的发展史")
inputEle.send_keys(Keys.ENTER)

# 等待5s
time.sleep(5)

# 获取当前网页的 URL
current_url = driver.current_url
print("-----------------------------------------------------------------------------------------------------------")
print("current_url:", current_url)

# 关闭浏览器
driver.quit()

# 解析 URL
parsed_url = urlparse(current_url)
path_parts = parsed_url.path.split('/')
conversation_id = path_parts[-1]
print("-----------------------------------------------------------------------------------------------------------")
print("conversation_id:", conversation_id)

url = 'https://api-search.tiangong.cn/chat-user/api/chat/conversation/detail?conversationId=' + conversation_id

print("请求地址：", url)

requests.get(url=url,timeout=10)

time.sleep(60)

response = requests.get(url=url,timeout=10)  # 超时10s

# 等待5s
time.sleep(5)

# 检查响应状态码
if response.status_code == 200:
    # 处理响应数据
    data = response.json()
    print("-----------------------------------------------------------------------------------------------------------")
    # print(data)
    # 使用 json.dumps 方法将 JSON 数据格式化为可打印的字符串，并指定缩进值以美化输出 指定 indent=4 参数，我们让输出的 JSON 字符串按照缩进四个空格的格式进行格式化输出
    print(json.dumps(data, ensure_ascii=False, indent=4))

else:
    print('请求失败，状态码：', response.status_code)
```
